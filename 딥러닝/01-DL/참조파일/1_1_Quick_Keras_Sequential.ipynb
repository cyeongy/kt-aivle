## Quick Linear Regression!

```
### 라이브러리 불러오기

import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20))
y = x * 2 -1 

print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API

# 1번. 세션 클리어 : 기존의 모델이 남아있다면 지워줘!
keras.backend.clear_session()

# 2번. 레이어 블록을 쌓을 발판 생성 : 레고처럼 차곡차곡!
model = keras.models.Sequential()

# 3번. 레이어 블록 조립
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1) )

# 4번. 컴파일
model.compile(loss='mse', optimizer='adam')

model.summary()
```

```
## 모델 학습

model.fit(x, y, epochs=10, verbose=1)
```

```
## 모델 예측

y_pred = model.predict(x)

y_pred[:5].reshape(-1)
```

```
# y_pred.shape
```

```
y[:5]
```

## Quick Logistic Regression!

```
## 라이브러리 불러오기

import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)

print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API

# 1번 세션 클리어
keras.backend.clear_session()

# 2번 레이어 블록 발판 생성 : 차곡차곡 쌓을거야
model = keras.models.Sequential()

# 3번 .add()
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

# 4번 .compile()
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])
```

```
## 모델 학습

model.fit(x, y, epochs=10, verbose=1)
```

```
## 모델 예측값 생성

model.predict(x)[:5].reshape(-1)
```

```
y[:5]
```

## Now, Your turn!

```
## 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,40)) 
y = np.array([0]*20 + [1]*20)
print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : 이진분류 Sequential API

## 세션 클리어 : 청소!
keras.backend.clear_session()

## 모델 발판 생성 : 레이어를 차곡차곡 쌓을!
model = keras.models.Sequential()

## 레이어 조립 : .add()
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

## 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy', keras.metrics.BinaryAccuracy(threshold=1.0)],
              optimizer='adam')
```

```
## 학습

model.fit(x, y, epochs=10, verbose=1)
```

```
## 예측값 생성
model.predict(x).reshape(-1)
```

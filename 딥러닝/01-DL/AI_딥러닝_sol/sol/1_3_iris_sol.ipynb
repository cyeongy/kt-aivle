```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
x = iris.data
y = iris.target
```

```
iris.target_names
```

```
# One-Hot Encoding
from tensorflow.keras.utils import to_categorical
```

```
y = to_categorical(y, 3)
```

```
x.shape, y.shape
```

# 1번. Sequential API style로 모델링 하세요

    loss = 'categorical_crossentropy' 사용!

```
import tensorflow as tf
from tensorflow import keras
```

```
# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(4,)) )
model.add( keras.layers.Dense(3, activation='softmax') )

# 오리지널 Sequential API
# model.add( keras.layers.Dense(3, input_shape=(4,), activation='softmax') )

# 4번 컴파일 
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
y_pred = model.predict(x)
```

# 2번. Functional API style로 모델링 하세요

```
x.shape, y.shape
```

```
import tensorflow as tf
from tensorflow import keras
```

```
# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 레이어 엮기
il = keras.layers.Input(shape=(4,))
ol = keras.layers.Dense(3, activation='softmax')(il)

# 3번. 모델의 시작과 끝 지정
model = keras.models.Model(il, ol)

# 4번. 모델 컴파일
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x)
```

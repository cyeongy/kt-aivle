# Keras로 Linear&Logistic Regression 조금 더 맛보기

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

# Linear Regression on 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 모델 선언
model = keras.models.Sequential()

# 3번. 모델 블록 조립
model.add( keras.layers.Input(shape=(13,)) )
model.add( keras.layers.Dense(1) )

# 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(13,)) )

# 4번. 모델 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x).reshape(-1)
```

# Logistic Regression 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_breast_cancer
```

```
import pandas as pd
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
```

```
print(cancer.DESCR)
```

```
pd.DataFrame(x, columns=cancer.feature_names).head()
```

```
y[:5]
```

```
cancer.target_names
```

```
x.shape, y.shape
```

```
# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 모델 선언
model = keras.models.Sequential()

# 3번. 모델 블록 조립
model.add( keras.layers.Input(shape=(30,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

# 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(30,), activation='sigmoid') )

# 4번. 모델 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x)
```

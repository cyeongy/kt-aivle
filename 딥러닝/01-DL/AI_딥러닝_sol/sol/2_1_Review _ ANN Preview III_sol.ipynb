```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
#####################
# 라이브러리 불러오기
#####################

import tensorflow as tf
from tensorflow import keras
```

```python
#######################
# 어떤 전처리 작업 필요
# Hint : y.shape
#######################

from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y, 3)

x.shape, y.shape
```

```python
########
# 모델링 I : Sequential
########

# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(4,)) )
model.add( keras.layers.Dense(3, activation='softmax') )

# 오리지널 Sequential API
# model.add( keras.layers.Dense(3, input_shape=(4,), activation='softmax') )

# 4번 컴파일 
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```python
########
# 모델링 II : Functional
########

# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 레이어 엮기
il = keras.layers.Input(shape=(4,))
ol = keras.layers.Dense(3, activation='softmax')(il)

# 3번. 모델의 시작과 끝 지정
model = keras.models.Model(il, ol)

# 4번. 모델 컴파일
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```python
###########
# 모델 학습
###########

model.fit(x, y, epochs=10, verbose=1)
```

```python
###########
# 모델 예측
###########

model.predict(x)
```

# 히든 레이어 추가하여 모델링! (Functional API)

    히든 레이어 2개를 추가하여 다시 모델링!
    히든 레이어 노드 수 : 32
    히든 레이어 activation = 'relu'

```python
########
# 모델링 I : Functional
########

# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 레이어 엮기
il = keras.layers.Input(shape=(4,))
hl = keras.layers.Dense(32, activation='relu')(il)
hl = keras.layers.Dense(32, activation='relu')(hl)
ol = keras.layers.Dense(3, activation='softmax')(hl)

# 3번. 모델의 시작과 끝 지정
model = keras.models.Model(il, ol)

# 4번. 모델 컴파일
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```python
########
# 모델링 II : Sequential
########

# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(4,)) )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(3, activation='softmax') )

# 오리지널 Sequential API
# model.add( keras.layers.Dense(32, activation='relu', input_shape=(4,)) )
# model.add( keras.layers.Dense(32, activation='relu') )
# model.add( keras.layers.Dense(3, activation='softmax') )

# 4번 컴파일 
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```python
###########
# 모델 학습
###########

model.fit(x, y, epochs=10, verbose=1)
```

```python
###########
# 모델 예측
###########

model.predict(x)
```
